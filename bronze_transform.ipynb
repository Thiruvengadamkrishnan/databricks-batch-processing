```python
# Base folder path
base_path = "/mnt/globalmart/"

# Load each CSV into a separate DataFrame
drivers_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(base_path + "drivers.csv")
locations_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(base_path + "locations.csv")
payments_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(base_path + "payments.csv")
trips_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(base_path + "trips.csv")
payment_methods_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(base_path + "payments.csv")
vehicles_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(base_path + "vehicles.csv")

display(drivers_df)
```

```python
%sql
-- Step 2: Create catalog and database for Bronze Layer (do not have access for creating catalog)
-- CREATE CATALOG IF NOT EXISTS globalmart_catalog;
CREATE DATABASE IF NOT EXISTS thiruvengadam_k_databricks_npmentorskool_onmicrosoft_com.bronzelayer1;
```

```python
# Step 3: Write to Bronze Layer Delta Tables
customers_df.write.format("delta").mode("overwrite").saveAsTable("thiruvengadam_k_databricks_npmentorskool_onmicrosoft_com.bronzelayer1.customers")

drivers_df.write.format("delta").mode("overwrite").saveAsTable("thiruvengadam_k_databricks_npmentorskool_onmicrosoft_com.bronzelayer1.drivers")

locations_df.write.format("delta").mode("overwrite").saveAsTable("thiruvengadam_k_databricks_npmentorskool_onmicrosoft_com.bronzelayer1.locations")

payments_df.write.format("delta").mode("overwrite").saveAsTable("thiruvengadam_k_databricks_npmentorskool_onmicrosoft_com.bronzelayer1.payments")

trips_df.write.format("delta").mode("overwrite").saveAsTable("thiruvengadam_k_databricks_npmentorskool_onmicrosoft_com.bronzelayer1.trips")

vehicles_df.write.format("delta").mode("overwrite").saveAsTable("thiruvengadam_k_databricks_npmentorskool_onmicrosoft_com.bronzelayer1.vehicles")



```
